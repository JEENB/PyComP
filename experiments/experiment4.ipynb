```python
symbols= ['0','1']
p = [0.1, 0.25, 0.35, 0.5, 0.01, 0.05]
dist = {"freq01" : [0.9, 0.1],
'freq025' : [75, 25],
'freq035' : [65, 35],
'freq05' :[1,1],
'freq001' : [0.99, 0.01],
'freq005' : [0.95, 0.05]}

symbol_len = [5,10,20,30, 50, 100, 150]

experiment_symbols = {}
for name, freq in dist.items():
    sym = []
    for slen in symbol_len:
        sym.append(get_symbols(symbols, freq, slen))
    experiment_symbols[name] = sym
```

```python
#freq 01
u = uABS(0.1)
ent = u.shannon_entropy()
msg = experiment_symbols['freq01']
comp_size01 = []
for m in msg:
    x, l = u.encode(m)
    comp_size01.append(len(bin(x))/l)
```
